# Logistics Regression

```python
#Importing Libraries
import sklearn 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

from sklearn.model_selection import RandomizedSearchCV
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler
# Using Standard scaler technique to convert all the numerical values into same decimal format
numerical_cols=['area', 'perimeter', 'compactness', 'length', 'width',
       'asymmetry coefficient', 'groove length']
scaler=StandardScaler()
scaled_cols=pd.DataFrame(scaler.fit_transform(df[numerical_cols]),columns=scaler.get_feature_names_out(numerical_cols))
```

```python
X=scaled_cols
y=df['category']
```

```python
X
```

```python

y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=5)
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
#train Score
Train_score = log.score(X_train,y_train)
print('Train_score: ',Train_score)
```

```python
#test Score
Test_score = log.score(X_test, y_test)
print('Test_Score: ',Test_score)
```

```python
pred_train = log.predict(X_train)
pred_test = log.predict(X_test)
```

```python
accuracy_logistics = accuracy_score(y_test,pred_test)
accuracy_logistics
```

```python
print(metrics.classification_report(y_test, pred_test))
```

```python
# Roc
```

```python
from sklearn.metrics import matthews_corrcoef

mcc= matthews_corrcoef(y_test,pred_test)
print("MCC: ",mcc)
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test,pred_test)
print('Confusion Matrix: ',)
print(cm)
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True, fmt = '.3g')
plt.title('Confusion Matrix - Test Data')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

# SVM

```python
#importing important libraries
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
from sklearn.metrics import roc_curve,auc
```

```python
svcm= SVC()
```

```python
#build the model
svcm.fit(X_train,y_train)
```

```python
#predicting the values of Y based on X-test
svcm_y_pred = svcm.predict(X_test)
```

```python
# accuracy
accuracy_SVM=accuracy_score(y_test,svcm_y_pred)
print('Accuracy:',accuracy_SVM)
```

```python
#classification report
print(classification_report(y_test,svcm_y_pred))
```

```python
#confusion matrix
cm = confusion_matrix(y_test,svcm_y_pred)
print('Confusion Matrix: ')
print(cm)
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True, fmt='.3g')
plt.title('Confusion Matrix - Test Data')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
#roc curve
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier

yb= label_binarize(y, classes=[0,1,2])
nc= yb.shape[1]
classifier = OneVsRestClassifier(SVC(kernel='linear', probability=True, random_state=0))
y_score = classifier.fit(X_train, y_train).decision_function(X_test)
fpr=dict()
tpr=dict()
roc_auc = dict()
for i in range(nc):
    fpr[i], tpr[i], _ = roc_curve(y_test ==i, y_score[:,i])
    roc_auc[i] = auc(fpr[i], tpr[i])
plt.figure()
color = ['blue','red','green']
for i, color in zip(range(nc),color):
    plt.plot (fpr[i], tpr[i], color=color, lw=2,
             label='ROC(area= {: .2f}) for class{}'.format(roc_auc[i],df['category'][i]))
plt.plot([0,1], [0,1], 'k--', lw=2)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC')
plt.show()
```
